```python
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt
df2 = pd.read_csv(filepath_or_buffer='./solar_production.csv', 
                  sep=';',
                  decimal='.')
rad=df2.copy()
rad.describe()
```

```python
x=df2["POA (W/m2)"]
y=df2["Pac (W)"]
hor = 10
ver = 4
plt.figure(figsize=(hor, ver))

plt.bar(x, y,label="Irradiance Vs Energy")

plt.rcParams['axes.axisbelow'] = True

plt.title('Irradiance POA (W/m2) Vs Pac-Energy (W)', fontsize=15)
plt.ylabel('Energy Pac (W)', fontsize=13)
plt.xlabel('Irradiance POA (W/m2)', fontsize=13)

plt.tick_params(direction='out', length=5, width=0.75, grid_alpha=0.3)
plt.xticks(rotation=90)
plt.ylim(None, None)
plt.xlim(None, None)
plt.grid(True)
plt.legend(loc='best', bbox_to_anchor=(1,1), fontsize=11)
plt.tight_layout;
plt.savefig('./bar_ver.png', dpi=400, bbox_inches='tight')
print("La grfica muestra una relacin creciente entre la irradiancia POA y la potencia AC entregada por el sistema, evidenciando que a mayor radiacin solar incidente, mayor es la potencia generada. En niveles bajos y medios de irradiancia, la relacin es casi lineal, lo que indica un buen aprovechamiento de la energa solar. Sin embargo, a partir de aproximadamente 1200 W/m se observan valores dispersos y fluctuaciones abruptas, posiblemente asociadas a la saturacin del inversor, variaciones rpidas de irradiancia por nubes o datos atpicos en la medicin. Este comportamiento sugiere que el sistema alcanza su lmite de potencia nominal y no incrementa su produccin a pesar de un aumento en la radiacin.")
```

```python
x=df2["Tmod (C)"]
y=df2["Pac (W)"]
hor = 10
ver = 4
plt.figure(figsize=(hor, ver))

plt.bar(x, y, label="Temperature vs Energy")

plt.rcParams['axes.axisbelow'] = True

plt.title('Tmod-Temperature (C) Vs Pac-Energy (W)', fontsize=15)
plt.ylabel('Energy Pac (W)', fontsize=13)
plt.xlabel('Temperature Tmod (C)', fontsize=13)

plt.tick_params(direction='out', length=5, width=0.75, grid_alpha=0.3)
plt.xticks(rotation=90)
plt.ylim(None, None)
plt.xlim(None, None)
plt.grid(True)
plt.legend(loc='best', bbox_to_anchor=(1,1), fontsize=11)
plt.tight_layout;
plt.savefig('./bar_ver_2.png', dpi=400, bbox_inches='tight')

print("La grfica presenta la relacin entre la temperatura del mdulo (`Tmod`) y la potencia AC (`Pac`) generada. Se observa una tendencia creciente, lo que indica que los valores ms altos de potencia coinciden con temperaturas de mdulo elevadas. Sin embargo, este comportamiento no necesariamente implica que el aumento de temperatura mejore el rendimiento, sino que ambas variables estn influenciadas por un mismo factor: la irradiancia solar. En das y horas de mayor radiacin, los mdulos alcanzan temperaturas ms altas y, simultneamente, producen ms potencia, aunque en condiciones reales un exceso de temperatura puede reducir ligeramente la eficiencia del mdulo fotovoltaico.")
```
